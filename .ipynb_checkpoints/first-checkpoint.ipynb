### 1.1) Basic encoding and decoding
Write a program that implements a neural representation of a scalar value  洧논 . For the neuron model, use a rectified linear neuron model ( 洧녩=洧녴洧녩洧논(洧냫,0) ). Choose the maximum firing rates randomly (uniformly distributed between 100Hz and 200Hz at x=1), and choose the x-intercepts randomly (uniformly distributed between -0.95 and 0.95). Use those values to compute the corresponding  洧띺  and  洧냫洧녪洧녰洧녩洧  parameters for each neuron. The encoders  洧  are randomly chosen and are either +1 or -1 for each neuron. Go through the following steps:

a). [1 mark] Plot the neuron responses  洧녩洧녰  for 16 randomly generated neurons. (See Figure 2.4 in the book for an example, but with a different neuron model and a different range of maximum firing rates).
Since you can't compute this for every possible  洧논  value between -1 and 1, sample the x-axis with  洧녬洧논=0.05 . Use this sampling throughout this question)

```python
%pylab inline
import numpy as np

class RLNeuron:
    
    def __init__(self):
        
        self.max_firing_rate = np.random.uniform(low = 100, high = 200)
        self.x_intercept = np.random.uniform(low = -0.95,high = 0.95)
        self.encoder = np.random.choice([1,-1])
        
        self.alpha = self.max_firing_rate/(1-self.x_intercept)
        self.j_bias = -self.alpha * self.x_intercept
        
        
def computeNeurons(n,X):
    neuron_values_array = np.zeros([n,len(X)])
    for i in range(0,n):
        neu = RLNeuron()
        for j,x in enumerate(X): 
            a = neu.encoder * neu.alpha * x + neu.j_bias
            neuron_values_array[i][j] = max(a, 0)
    return neuron_values_array


number_of_samples = (1-(-1))/0.05
number_of_neurons = 16
X = np.linspace(-1,1,number_of_samples)
A = computeNeurons(number_of_neurons, X)

for i in range(0,number_of_neurons):
    plot(X, A[i])
xlabel('$x$ (stimuli)')
ylabel('$a$ (firing rate)')
title('Rectified Linear Neurons');
```

b). [1 mark] Compute the optimal decoders  洧녬洧녰  for those 16 neurons (as shown in class). Report their values.
The easiest way to compute  洧녬  is to use the matrix notation mentioned in the course notes.  洧냢  is the matrix of neuron activities (the same thing used to generate the plot in 1.1a).

```python
upsilon = np.matmul(A,X) / number_of_samples
gamma = np.matmul(A, np.transpose(A))/ number_of_samples
gamma_inv = np.linalg.inv(gamma)
d = np.dot(gamma_inv, upsilon)
print(d)
```

c). [1 mark] Compute and plot  洧논팩 =갬洧녰洧녬洧녰洧녩洧녰 . Overlay on the plot the line  洧녽=洧논 . (See Figure 2.7 for an example). Make a separate plot of  洧논뉧롐봛   to see what the error looks like. Report the Root Mean Squared Error value.

```python
X_hat = np.matmul(np.transpose(A),d)

def getRMS(X, X_hat):
    RMS = 0
    Diff = X - X_hat
    for x in range(0,len(X)):
        RMS += np.power(Diff[x],2)
    RMS = np.sqrt(RMS/len(X))
    return RMS, Diff

RMS1, Diff = getRMS(X, X_hat)
print("RMS error = ", str(RMS1))


```
